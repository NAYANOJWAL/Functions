## **THEORY QUESTIONS**


**1. What is the difference between a function and a method in Python?**
  - A **function** is an independent block of code.
  Ex: len([1,2,3]) -> is a function

  - A **method** is a function associated with an object (called with dot notation).
  Ex: [1,2,3].append(4) -> is a method.


**2. Explain the concept of function arguments and parameters in Python.**
  - **Parameters** are variables listed in a function definition.
  Ex: def add(a, b): -> a, b are parameters

  - **arguments** are values passed to the function during a call.
  Ex: add(5, 3) -> 5, 3 are arguments.


**3. What are the different ways to define and call a function in Python?**
  - Functions can be defined using **def** or **lambda**, and called by writing the function name with arguments.
  Ex: def greet(): print("Hello") -> greet();
  mul = (lambda x: x*2) -> mul(3).


**4. What is the purpose of the return statement in a Python function?**
  - The return statement exits a function and send result back to the caller.
  Ex: def square(x): return x*x -> square(4) returns 16.


**5. What are iterators in Python and how do they differ from iterables?**
  - An iterable is any object you can loop over
  Ex: nums = [1, 2, 3] is iterable.

  - an iterator is an object with a __next__() method created using iter().
  Ex: it = iter(nums) is an iterator -> next(it) gives 1.


**6. Explain the concept of generators in Python and how they are defined.**
  - Generators are functions that yield values one at a time using the yield keyword, producing an iterator.
  Ex: ef count(): yield 1; yield 2 -> for i in count(): print(i) prints 1 and 2.


**7. What are the advantages of using generators over regular functions?**
  - Generators use less memory and support lazy evaluation, making them efficient for large or infinite sequences.
  Ex: def infinite(): n = 0; while True: yield n; n += 1 gives an infinite number stream without memory overload.



**8. What is a lambda function in Python and when is it typically used?**
  - A lambda is an anonymous, a consie way to write a function used for short-term operation defined in single-line.
  Ex: square = lambda x: x**2 -> square(4) returns 16.



**9. Explain the purpose and usage of the map() function in Python.**
  - The map() function applies a given function to all items in an iterable and returns a map object.
  Ex: map(str.upper, ['a', 'b']) -> returns ['A', 'B']



**10. What is the difference between map(), reduce(), and filter() functions in Python?**
  - map() transforms items
  Ex: map(lambda x: x*2, [1,2]) -> [2,4]

  -filter() selects items
  Ex: filter(lambda x: x%2==0, [1,2]) -> [2]

  - reduce() aggregates items into a single value
  Ex: reduce(lambda x,y: x+y, [1,2,3]) -> 6

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even(num):
  sum = 0
  for i in num:
    if(i%2==0):
      sum = sum + i
  return sum

print(sum_even([1,2,3,4,5,6]))
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def rev_string(s):
  r = ''
  for i in s:
    r = i + r
  return r

print(rev_string('hello'))
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square(lst):
  lst1 = []
  for i in lst:
    lst1.append(i**2)
  return lst1

print(square([1,2,3,4]))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
  if n<2:
    return False
  for i in range(2,n):
    if(n%i==0):
      return False
  return True

for i in range(1,201):
  if(prime(i) == True):
    print(i, end = ' ')
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class fib:
    def __init__(self, terms): self.n, self.a, self.b = terms, 0, 1
    def __iter__(self): return self
    def __next__(self):
        if self.n <= 0: raise StopIteration
        self.a, self.b, self.n = self.b, self.a + self.b, self.n - 1
        return self.a
print(list(fib(7)))
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power(e):
  for i in range(e+1):
    yield 2 ** i

print(list(power(5)))
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
with open('sample.txt', 'w') as f:
  f.write('Hello world\nMy name is nayanojwal sri tej\nThank you')

def read(name):
  with open(name) as n:
    for line in n:
      yield line.strip()

for line in read('sample.txt'):
  print(line)


```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tup = [(1,3), (2,1), (4,2)]
sort_tup = sorted(tup, key = lambda x: x[1])
print(sort_tup)
```

```python
# 9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
cel = [0,20,37,100]
far = list(map(lambda x: (x*9/5)+32, cel))
print(far)
```

```python
# 10. Create a Python program that uses filter() to remove all the vowels from a given string.
s = 'nayanojwal'
no_vowel = ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
print(no_vowel)
```
